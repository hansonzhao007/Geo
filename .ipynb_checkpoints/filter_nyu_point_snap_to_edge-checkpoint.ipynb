```python
import folium
from folium import plugins
import pandas as pd
import random
import json
# latitude, longitude
location = [40.783939, -73.979752]
m = folium.Map(location,tiles='Stamen Toner', zoom_start=11)
```

```python
geo = r'GeojsonGraphData/nyu_point_link_edge.json'

#Let Folium determine the scale
roads = folium.GeoJson(
    geo,
    style_function=lambda feature: {
        'fillColor': '#00ff00',
        'color': 'red',
        'weight': 2,
        'dashArray': '5, 5',
        'fillOpacity': 0.5,
    }
)
m.add_child(roads)

# m.save('nyu_with_roads.html')
```

```python
geo = r'GeojsonGraphData/filter_nyu_taxi.json'

with open(geo) as f:
        data = json.load(f)
df = []
for feature in data['features']:
    df.append(feature["geometry"]["coordinates"]);
df = pd.DataFrame(df,columns=['lng','lat']) 
heat_data = [[row['lat'],row['lng']] for index, row in df.iterrows()]

heat_map = plugins.HeatMap(heat_data, 
                           radius = 7,
                           min_opacity = 0.8,
                           blur=2,
                           #max_zoom=1
                          )
m.add_child(heat_map)
```

```python
m.save('nyu_nearest_edge.html')
```

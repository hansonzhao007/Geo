```python
import folium
from folium import plugins
import pandas as pd
import random
```

```python
p = 0.01
data = pd.read_csv('./cellid.csv', 
                   nrows=10000, 
                   skiprows=lambda i: i>0 and random.random() > p)
print(data.sample(10))
print(data.count())
longitude_lim=(data['longitude'].min(),data['longitude'].max())
latitude_lim=(data['latitude'].min(),data['latitude'].max())
print(longitude_lim, latitude_lim)
```

```python
m = folium.Map(location=[40.830429, -73.891762], zoom_start=11)
m
```

```python
data['latitude'] = data['latitude'].astype(float)
data['longitude'] = data['longitude'].astype(float)

heat_data = [[row['latitude'],row['longitude']] for index, row in data.iterrows()]

m = folium.Map([40.830429, -73.891762], zoom_start=11)
heat_map = plugins.HeatMap(heat_data, 
                           radius = 7,
                           min_opacity = 0.6,
                           blur=15,
                           #max_zoom=1
                          )
m.add_child(heat_map)
m
```

```python
m.save('./heat_map.html')
```
